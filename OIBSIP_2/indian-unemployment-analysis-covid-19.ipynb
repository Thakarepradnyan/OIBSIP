Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly .express as px
```

loading the dataset

```python
data = pd.read_csv('/kaggle/input/covid-unemployment-india/Unemployment in India.csv')
data = pd.read_csv('/kaggle/input/covid-unemployment-india/Unemployment_Rate_upto_11_2020.csv')
print(data.head(5))
```

```python
#checking for duplicates
print(data.isnull().sum())
```

Defining data columns

```python
data.columns = ['States','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Region','Longitude','Latitude']
```

Checking the corelation between future of dataset

```python
# plotting correlation heatmap

import warnings
from matplotlib.cbook import MatplotlibDeprecationWarning

# Suppress warnings
warnings.filterwarnings("ignore", category=MatplotlibDeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10,8))
sns.heatmap(data.corr())
plt.show()
```

Data visualization 

Estimated no of employee according to different region of india

```python
# plotting histplot

data.columns=['States','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Region','Longitude','Latitude']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

Unemployment rate according to different regions of india

```python
# plotting histplot

plt.figure(figsize=(10,8))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue='Region',data=data)
plt.show()
```

Dashboard to analyze the unemployment rate of each Indian state

```python
# plotting sunburst

unemployment = data[['States','Region','Estimated Unemployment Rate']]
figure = px.sunburst(unemployment,path=['Region','States'],
                     values='Estimated Unemployment Rate',
                     width=700,height=600, color_continuous_scale='RdY1Gn',
                     title="Unemployment Rate in India")
figure.show()
```
